```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv('../01. Scraping/raw_data.csv')
print(df)
```

### Brand names to be extracted

```python
brands = [
    'asus', 'dell', 'lenovo', 'zyrex', 'hp', 'acer', 'apple', 'msi', 'samsung',
    'toshiba', 'huawei', 'xiaomi', 'razer', 'lg', 'sony', 'alienware', 'tecno',
    'hasee', 'fujitsu', 'gigabyte', 'surface', 'axioo', 'advan', 'google', 'infinix',
    'ast', 'maxone', 'vaio', 'microsoft', 'amoli', 'chuwi' , 'spc'
]
```

```python
df['Name'] = (
    df['Name']
    .str.replace('thinkpad', 'lenovo', case=False)
    .str.replace('macbook', 'apple', case=False)
    .str.replace('yoga', 'lenovo', case=False)
    .str.replace('max one', 'maxone', case=False)
    .str.replace('thosiba', 'toshiba', case=False)
    .str.replace('m1cro5oft', 'microsoft', case=False)
    .str.replace('x360', 'hp', case=False)
    .str.replace('axxioo', 'axioo', case=False)
    .str.replace('ideapad', 'lenovo', case=False)
    .str.replace('l380', 'lenovo', case=False)
    .str.replace('t470', 'lenovo', case=False)
    .str.replace('latitude', 'dell', case=False)

)
```

### Extract brand names

```python
df['Name'] = (
    df['Name']
    .str.findall(rf'(?i)({"|".join(brands)})')
    .str[0]
    .str.capitalize()
)
df
```

### Save the brandless data

```python
nan_df_brand = df[df['Name'].isna() | (df['Name'] == False)]
nan_df_brand.to_csv('brandless_data.csv')
```

```python
df = df.rename(columns={'Name':'Brand'})
df = df.dropna(subset='Brand')
```

```python
df['Rating'] = df['Rating'].replace(np.nan, '0')
df['Sold'] = df['Sold'].replace(np.nan, '0 terjual')
```

### Save the cleaned data

```python
df.to_csv('clean_data.csv', index=False)
```

### Format the price column

```python
df['Price'] = (
    df['Price']
    .str.replace('Rp', '')
    .str.replace('.', '')
)

df['Price'] = pd.to_numeric(df['Price'], errors='coerce')
df = df.rename(columns={'Price':'Price (Rp)'})
df
```

### Format the sold column

```python
df['Sold'] = (
    df['Sold']
    .str.replace('rb', '000')
    .str.replace('+', '')
    .str.replace('terjual', '')
)

df['Sold'] = pd.to_numeric(df['Sold'], errors='coerce')
df['Sold'] = df['Sold'].astype('Int64')
df
```

### Save the formatted data

```python
df.to_csv('formatted_data.csv', index=False)
```
